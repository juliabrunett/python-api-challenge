```python
# Import dependencies
import matplotlib.pyplot as plt
import requests
import pandas as pd
import numpy as np
from config import gkey
import gmaps
import os
```

```python
# Import file from Part 1, and read into a dataframe
weather_data_df = pd.read_csv("../WeatherPy/OutputData/cities_weather_data.csv")
weather_data_df
```

```python
# Add city lat and long to a new df
cities = weather_data_df[["Latitude", "Longitude"]].astype(float)

# Add humidity to a new df
humidity = weather_data_df["Humidity"].astype(float)
```

```python
# Configure gmaps
gmaps.configure(api_key=gkey)
```

```python
# Configure gmaps
gmaps.configure(gkey)
```

```python
# Create heatmap for data depending on humidity
figure_layout = {
    'width': '600px',
    'height': '500px',
    'border': '1px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}

# Create a gmaps figure
fig = gmaps.figure(layout=figure_layout)

heat_layer = gmaps.heatmap_layer(cities, weights=humidity, 
                                 dissipating=True)

fig.add_layer(heat_layer)
fig
```

```python
# Change max intensity and point radius of the map
heat_layer.max_intensity = 100
heat_layer.point_radius = 15
```
